#### Method 1 - Performing groupby seperately and then merging the columns

```python
groupby_quantity = demand_transact.groupby(['City', 'Mapped_Sales_Type', 'Mapped_Item_Code', 'Transaction_Date'])['Quantity_Sold'].aggregate('sum')
groupby_quantity
```

```python
groupby_medianprice = demand_transact.groupby(['City', 'Mapped_Sales_Type', 'Mapped_Item_Code', 'Transaction_Date'])['Median_Price'].aggregate('median')
groupby_medianprice
```

```python
groupby_effectiveprice = demand_transact.groupby(['City', 'Mapped_Sales_Type', 'Mapped_Item_Code', 'Transaction_Date'])['Effective_Price'].aggregate('median')
groupby_effectiveprice
```

```python
pd.merge(pd.merge(groupby_quantity,groupby_medianprice,right_index=True,left_index=True),groupby_effectiveprice,right_index=True,left_index=True)
```

### Add a column “Range” to the existing dataframe for below condition
#### a. 1 for Effective_Price > 500
#### b. 0.5 otherwise

### Approach1 - using the traditional if-else condition

```python
counter = 0
for values in demand_transact['Effective_Price']:
    if values > 500:
        demand_transact.loc[counter,'Range']=1
    else:
        demand_transact.loc[counter,'Range']=0.5
    counter = counter + 1
```

```python
demand_transact[['Effective_Price','Range']]
```

```python
demand_transact['Effective_Price'].dtype
```

```python
demand_transact.loc[0,'Effective_Price']
```

### Approach2 - Using the lambda function

```python
demand_transact['Range1'] = demand_transact['Effective_Price'].apply(lambda x : 1 if x>500 else 0.5 )
```

#### Handle the missing values in the column Percent_Quantity_With_Discount with strategy = mean

```python
demand_transact1[demand_transact1['Percent_Quantity_With_Discount'].isna()]
```

```python
demand_transact1['Percent_Quantity_With_Discount'].isna().sum()
#There are in total 18022 records that have NA values in the dataframe.
```

```python
##Percentage of NA rows in the feature 'Percent_Quantity_With_Discount'
demand_transact1['Percent_Quantity_With_Discount'].isna().sum()/len(demand_transact1['Percent_Quantity_With_Discount'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(demand_transact1['Percent_Quantity_With_Discount'])
```

```python
sns.boxplot(x='Percent_Quantity_With_Discount',data=demand_transact1)
```

```python
demand_transact1['Percent_Quantity_With_Discount'].describe()
```

```python
#In this feature we have around 60% of the data that have NA values 
#The mean imputation does not seem to be good option, since the dataset seems to have many outliers in it. 
```

```python
mean_val = np.mean(demand_transact1['Percent_Quantity_With_Discount'])
demand_transact1['New_Col'] = demand_transact1['Percent_Quantity_With_Discount'].fillna(value=mean_val)
```

```python
demand_transact1[['Percent_Quantity_With_Discount','New_Col']]
```

```python
sns.FacetGrid(demand_transact1,col=)
```

```python
#Plotting the Percent_Quantity_With_Discount vs New_Col after Mean imputation

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].set_title('Before Mean Imputation')
axes[1].set_title('After Mean Imputation')
sns.histplot(ax=axes[0],data=demand_transact1,x='Percent_Quantity_With_Discount')
sns.histplot(ax=axes[1],data=demand_transact1,x='New_Col')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].set_title('Before Mean Imputation')
axes[1].set_title('After Mean Imputation')
sns.boxplot(ax=axes[0],data=demand_transact1,x='Percent_Quantity_With_Discount')
sns.boxplot(ax=axes[1],data=demand_transact1,x='New_Col')
plt.show()
# sns.boxplot(demand_transact1['New_Col'])
```

```python
#Performing median imputation
median_val = np.nanmedian(demand_transact1['Percent_Quantity_With_Discount'])
# print(np.median(demand_transact1['Percent_Quantity_With_Discount'])) - The np.median() function returns nan
# print(np.nanmedian(demand_transact1['Percent_Quantity_With_Discount']))
print('The median value is :',median_val)
demand_transact1['New_Col'] = demand_transact1['Percent_Quantity_With_Discount'].fillna(value=median_val)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].set_title('Before Median Imputation')
axes[1].set_title('After Median Imputation')
sns.histplot(ax=axes[0],data=demand_transact1,x='Percent_Quantity_With_Discount')
sns.histplot(ax=axes[1],data=demand_transact1,x='New_Col')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].set_title('Before Median Imputation')
axes[1].set_title('After Median Imputation')
sns.boxplot(ax=axes[0],data=demand_transact1,x='Percent_Quantity_With_Discount')
sns.boxplot(ax=axes[1],data=demand_transact1,x='New_Col')
plt.show()
```
